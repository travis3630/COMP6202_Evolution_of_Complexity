```python
import random 
import string
import numpy

string = 'methinks it is like a weasel'
```

```python
###   1) Representing and initialising individuals    ###
def random_string(length):
  ran_string = ''
  for i in range(length):
    ran_string += random.choice(chr(random.randrange(32, 127)))
  #print(ran_string)
  return ran_string
rans = random_string(len(string))
print('random string = ' + rans )
```

```python
###   2) Evaluating individuals     ###
def fitness(original,offspring):
  score = 0
  for i in range(len(original)):
    if(original[i]==offspring[i]):
      score+=1
  return score
```

```python
###   3) Mutation   ###
def mutation(original):
  prob = 1.000/len(original)
  original = list(original)
  for i in range(len(original)):
    if(random.uniform(0,1)< prob):
      original[i] = random.choice(chr(random.randrange(32, 127)))
  original = ''.join(map(str,original))
  return original
```

```python
###   4) A mutation hill-climber    ###
def mut_hill_climber(original, parent):
  fit_score = fitness(original,parent)
  child = ''
  while(fit_score < len(original)):
    child = mutation(parent)
    if(fitness(original,child) > fitness(original,parent)):
      parent = child
      fit_score = fitness(original,parent)
    print(parent + ' , fitness= ' + str(fitness(original,parent)))
  return parent
```

```python
rans = random_string(len(string))
score = fitness(string,rans)

rans = mut_hill_climber(string, rans)
```

```python
###   4) A GA without crossover     ###
def GA_wo_crossover(original,population):
  pop_fitness = []
  fitness_goal = len(original)
  fitness_score_child = 0
  for i in range(len(population)):
    pop_fitness.append(fitness(original,population[i]))
  while(fitness_score_child != len(original)):
    
    A = random.randint(0, len(population)-1)
    B = random.randint(0, len(population)-1)
    if(fitness(original,population[A])>fitness(original,population[B])):
      parent = population[A]
    else:
      parent = population[B]
    child = mutation(parent)#Mutate parent -> child
    fitness_score_child = fitness(original,child)
    C = random.randint(0, len(population)-1)
    D = random.randint(0, len(population)-1) 
    if(fitness(original,population[C])>fitness(original,population[D])):
      population[D] = child
    else:
      population[C] = child
    print(fitness_score_child, end= ', ')
  return population
```

```python
population = []
for i in range(500):
  population += [random_string(len(string))]
population = GA_wo_crossover(string,population)
print(population)
```

```python
###   5) Crossover.   ###
def crossover(parent1,parent2):
  child = ''
  for i in range(len(parent1)):
    if(random.uniform(0,1)<0.5):
      child += parent1[i]
    else:
      child += parent2[i]
  return child
```

```python
parent1 = random_string(28)
parent2 = random_string(28)
child = crossover(parent1,parent2)
print(parent1)
print(parent2)
print(child)
```

```python
###   6) A GA with crossover    ###
def GA_with_crossover(original,population):
  pop_fitness = []
  fitness_goal = len(original)
  fitness_score_child = 0
  for i in range(len(population)):
    pop_fitness.append(fitness(original,population[i]))
  while(fitness_score_child != len(original)):
    A = random.randint(0, len(population)-1)
    B = random.randint(0, len(population)-1)
    if(fitness(original,population[A])>fitness(original,population[B])):
      parent1 = population[A]
    else:
      parent1 = population[B]
    C = random.randint(0, len(population)-1)
    D = random.randint(0, len(population)-1)
    if(fitness(original,population[C])>fitness(original,population[D])):
      parent2 = population[C]
    else:
      parent2 = population[D]
    child = crossover(parent1,parent2)# crossover parent
    child = mutation(child)                  #Mutate parent -> child
    fitness_score_child = fitness(original,child)
    E = random.randint(0, len(population)-1)
    F = random.randint(0, len(population)-1) 
    if(fitness(original,population[E])>fitness(original,population[F])):
      population[F] = child
    else:
      population[E] = child
    print(fitness_score_child, end = ', ')
  return population
```

```python
population = []
for i in range(500):
  population += [random_string(len(string))]
population = GA_with_crossover(string,population)
```
